```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
file_path = "/content/dataset.csv"
df = pd.read_csv(file_path, skiprows=4)  # Skipping extra header rows

# Step 2: Select the latest available population data
df = df[['Country Name', 'Country Code', '2022']]  # Use latest available year
df = df.rename(columns={'2022': 'Population'})  # Rename column for clarity

# Step 3: Remove missing values
df = df.dropna()

# Step 4: Sort by Population (Top 10 for better visualization)
df_sorted = df.sort_values(by="Population", ascending=False).head(10)

# Step 5: Create a Bar Chart (Top 10 Most Populated Countries)
plt.figure(figsize=(12, 6))
sns.barplot(x="Population", y="Country Name", data=df_sorted, palette="viridis")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.title("Top 10 Most Populated Countries in 2022")
plt.grid(axis='x', linestyle="--", alpha=0.7)
plt.show()

# Step 6: Create a Histogram (Distribution of Population)
plt.figure(figsize=(10, 5))
sns.histplot(df["Population"], bins=20, kde=True, color="blue")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population Across Countries")
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.show()
```
